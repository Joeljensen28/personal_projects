```python
from bs4 import BeautifulSoup
import requests

website = requests.get('https://en.wikipedia.org/wiki/List_of_members_of_the_Quorum_of_the_Twelve_Apostles_(LDS_Church)')
soup = BeautifulSoup(website.text, 'html.parser')
```

```python
# Find all tables on the page that have data on apostles
tables = soup.find_all('table', {'class': 'wikitable'})

# Pluck each table and store it in separate lists
current = tables[0]
original = tables[1]
nineteenth_century = tables[2]
twentieth_century = tables[3]
twentyfirst_century = tables[4]
never_members = tables[5]
clark = tables[6]
```

```python
# Split the lists on `\n` and compile them all into the same list
data = current.get_text().split('\n')
data.extend(original.get_text().split('\n'))
data.extend(nineteenth_century.get_text().split('\n'))
data.extend(twentieth_century.get_text().split('\n'))
data.extend(twentyfirst_century.get_text().split('\n'))
data.extend(never_members.get_text().split('\n'))
data.extend(clark.get_text().split('\n'))

# Remove all the items that have empty space
new = list(filter(lambda x: x != '', data))
```

```python
for i in range(len(new)):
    # Loop through each item and remove the data on notes and death
    if new[i] == 'Notes:' or new[i] == 'Died:':
        # Remove the 'Notes:' or 'Died' headers
        new.pop(i)
        # Followed by removing the next immediate item that contains the actual data
        new.pop(i)
# Note that this loop will terminate in an error since the list has shrunk in size since the initialization of the for loop, but the purpose is accomplished.
```

```python
# Remove all headers
clean_list = list(filter(lambda x: not x.endswith(':'), new))
## Note that filter isn't used in the previous cell because I wanted to remove the data associated with Notes: and Death: and not just the header
```

```python
# The wikipedia page had inconsistent formatting, resulting in a few interval errors in cell 30. This line takes care of those cases.
clean_list = list(filter(lambda x: x not in ['Uncle of N. Eldon Tanner', 'Nephew of Hugh B. Brown'], clean_list))
```

```python
import pandas as pd

df = pd.DataFrame()

# Make some lists that can be plugged into df as columns once they're populated
name = []
birth = []
pos = []

for i in range(len(clean_list)):
    # Get the current item
    item = clean_list[i]
    # This sequence checks if the item is a name, birthday, or position
    if i % 3 == 0:
        name.append(item)
    elif i % 3 == 1:
        birth.append(item)
    else:
        pos.append(item)

# Add the collected data to the df accordingly
df['name'] = name
df['birthdate'] = birth
df['positions'] = pos
```

```python
df.to_csv('apostles_uncleansed.csv')
```
